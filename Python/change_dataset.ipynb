```python
# --- import external libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import os
import math
import cv2
```

```python
# --- Loading dataset
input_folder_path = "./data/"
train_df = pd.read_csv(input_folder_path+"train.csv")
test_df = pd.read_csv(input_folder_path+"test.csv")
```

```python
df_train_0 = train_df[train_df.label == 0]
df_train_1 = train_df[train_df.label == 1]
df_train_5 = train_df[train_df.label == 5]
df_train_9 = train_df[train_df.label == 9]
```

```python
result = pd.concat([df_train_0 ,df_train_1 ,df_train_5 ,df_train_9])
```

```python
result
```

```python
result_df = pd.DataFrame()
for i in range(len(result)):
    temp = np.array(result.iloc[i, 1:])
    temp = temp.reshape(28, 28)
    temp = temp.astype(np.uint8)
    temp = cv2.resize(temp, (10,10))
    temp = temp.reshape(10*10)
    dataframe = np.concatenate((np.array([result.iloc[i,0]]), temp))
    result_df = result_df.append(pd.DataFrame(dataframe).T)
```

```python
result_df
```

```python
result_df_new = result_df.rename(columns={0: 'label'})
result_df_new
```
